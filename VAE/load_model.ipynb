```python
%matplotlib inline
import torch
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np

def show(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
```

```python
with open("models/VAE", "rb") as f:
    my_model = torch.load(f)
```

```python
image_examples = my_model.generate(torch.autograd.Variable(torch.randn(15 * 15, 30).cuda())).view(-1, 1, 28, 28)
```

```python
image_grid = make_grid(image_examples.data, nrow=15).cpu()

plt.figure(figsize=(10, 10))
show(image_grid)
```
